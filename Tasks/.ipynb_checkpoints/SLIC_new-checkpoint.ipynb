```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.image import imread
```

```python
def get_img(img_path, show=False):
    orig_img = imread(img_path)
    if show:
        plt.imshow(orig_img)
        plt.show()
        print('Image shape: ', orig_img.shape)
    return orig_img


def image_show(before, after):
    fig, axarr = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(20,8))
    axarr[0].imshow(before)
    axarr[1].imshow(after.astype(np.uint8))
    plt.show()
```

```python
def get_kmeans(clusters, n_colors):
    kmeans = KMeans(n_clusters=n_colors).fit(clusters)
    sq = kmeans.predict(clusters)
    # [print(i) for i in sq]
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    return centroids[labels].reshape((-1, 5))

```

```python
orig_img = get_img('res\medved.jpg')

image = orig_img
img_height = image.shape[0]
img_width = image.shape[1]

S = 100
h = 0
w = 0

clusters = []
while h < img_height:
    while w < img_width:
        element = [image[h, w][0], image[h, w][1], image[h, w][2], h, w]
        clusters.append(element)
        w += 1
    w = 0
    h += 1

new_img_ = get_kmeans(clusters, S)[:, [0, 1, 2]]
new_image = new_img_.reshape(orig_img.shape)

image_show(orig_img, new_image)
```
