```python
import pygame
import numpy as np
import matplotlib.pyplot as plt

RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
    
data = np.empty((0,3), dtype='f')
class_ = 1

def create_data(position):
    (x, y) = position
    r = np.random.uniform(0,30)
    phi = np.random.uniform(0,2*np.pi)
    global class_
    color = class_
    point = [x + r * np.cos(phi), y + r * np.sin(phi), color]
    global data
    data = np.append(data, [point], axis=0)


radius = 2
colors = {1: RED, 2: GREEN, 3: BLUE}
thickness = 0
    
bg_color = (255, 255, 255)
(width, height) = (640, 480)

screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("TEST")

running = True
pushing = False
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            pushing = True
        elif event.type == pygame.MOUSEBUTTONUP:
            pushing = False
        elif event.type == pygame.KEYDOWN:
            if event.unicode == 'r':
               class_ = 1
            elif event.unicode == 'g':
               class_ = 2
            elif event.unicode == 'b':
               class_ = 3

    if pushing and np.random.uniform(0,1)>.5:
        create_data(pygame.mouse.get_pos())
        
    screen.fill(bg_color)
    for point in data:
        pygame.draw.circle(screen, colors.get(point[2]), (int(point[0]), int(point[1])), radius, thickness)
    pygame.display.flip()
pygame.quit()
```

```python
X = data[:,0:2]
bias = np.array([[-1 for i in range(X.shape[0])]])
X = np.append(X, bias.T, axis=1)

y = data[:,2]
y[y > 1.] = -1.

plt.figure()
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
from sklearn.utils import shuffle

def svm_sgd(X, y):
    w = np.zeros(X.shape[1])
    epochs = 680
    
    for epoch in range(1,epochs):
        h = 1/epoch
        X_, y_ = shuffle(X, y)
        for i,x in enumerate(X_):
            dist = 1 - y_[i]*(np.dot(X_[i],w))
            if max(0, dist) == 0:
                w[:-1] -= h*(w[:-1])/len(X_)
            else:
                w[:-1] -= h*(w[:-1] - X_[i][:-1]*y_[i])/len(X_)
                w[-1] -= h*y_[i]
    return w

w = svm_sgd(X, y)
y_test = np.sign(np.dot(X, w))

plt.figure()
plt.scatter(X[:,0], X[:,1], c=y_test, cmap=plt.cm.Paired)
plt.plot([0, w[-1]/w[0]], [w[-1]/w[1], 0], marker='s')
plt.show()
```
