```python
import pygame
import numpy as np

RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
YELLOW = (255, 255, 0)


class Point:
    def __init__(self, x=0, y=0, color=YELLOW):
        self.color = color
        self.x = x
        self.y = y
    
    
    def get_color(self):
        return self.color
    
    
    def get_x(self):
        return self.x
    
    
    def get_y(self):
        return self.y
    
    
data = list()
class_ = YELLOW

def create_data(position):
    (x, y) = position
    r = np.random.uniform(0,30)
    phi = np.random.uniform(0,2*np.pi)
    global class_
    color = class_
    point = Point(x + r*np.cos(phi), y + r*np.sin(phi), color)
    global data
    data.append(point)


radius = 2
thickness = 0
    
bg_color = (255, 255, 255)
(width, height) = (640, 480)

screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("TEST")

running = True
pushing = False
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            pushing = True
        elif event.type == pygame.MOUSEBUTTONUP:
            pushing = False
        elif event.type == pygame.KEYDOWN:
            if event.unicode == 'r':
               class_ = RED
            elif event.unicode == 'g':
               class_ = GREEN
            elif event.unicode == 'b':
               class_ = BLUE
            elif event.unicode == 'y':
               class_ = YELLOW
            
    if pushing and np.random.uniform(0,1)>.5:
        create_data(pygame.mouse.get_pos())
        
    screen.fill(bg_color)
    for point in data:
        pygame.draw.circle(screen, point.get_color(), (int(point.get_x()), int(point.get_y())), radius, thickness)
    pygame.display.flip()
pygame.quit()
```
