```python
import matplotlib.pyplot as plt
import skimage.segmentation as seg
from skimage import io
from skimage import color


def image_show(image):
    fig, axarr = plt.subplots(nrows=1, ncols=1, sharex=True, figsize=(20,8))
    axarr.imshow(image)

    
orig_image = io.imread('D:/Desktop/w.png')
#orig_image = io.imread('D:/Desktop/6.jpg')
slic_image = seg.slic(orig_image, n_segments=100)
image_show(color.label2rgb(slic_image, orig_image, kind='avg'));
```
